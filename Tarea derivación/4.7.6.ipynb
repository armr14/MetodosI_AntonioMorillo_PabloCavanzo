```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4, 4, 25)
y = np.linspace(-4, 4, 25)
x, y = np.meshgrid(x, y)
h = 0.001
Vx = 2
R = 2
def f(x,y):
  return Vx*x*(1 - ((R**2)/((x**2) + (y**2))))
def CentralDerivativex(fn,x,h,y):
  d = (fn(x+h,y) - fn(x-h,y))/(2*h)
  return d

def CentralDerivativey(fn,x,h,y):
  d = (fn(x,y+h) - fn(x,y-h))/(2*h)
  return d

vx = CentralDerivativex(f,x,h,y)
vy = -1*CentralDerivativey(f,x,h,y)
vx[x**2 + y**2 <= R**2] = 0
vy[x**2 + y**2 <= R**2] = 0
fig, ax = plt.subplots()
ax.quiver(x, y, vx, vy, color='b')
ax.set_aspect('equal')
plt.xlim(-4,4)
plt.ylim(-4,4)
patch = plt.Circle((0,0),R,fill=False, color = 'r')
ax.add_patch(patch)
plt.show()
```
